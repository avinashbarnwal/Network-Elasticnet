```python
import numpy    as np
import pandas   as pd
import os
import seaborn  as sns;
import matplotlib.pyplot as plt
from   numpy    import linalg as LA
```

```python
data                = pd.read_csv("../../result/final_result.csv")
data['TUMOR_STAGE'] = data['TUMOR_STAGE'].apply(lambda x : 'Stage IIIA' if pd.isnull(x) else x)
data.set_index('Index',inplace=True)
data.sort_index(inplace=True)
```

```python
data_t       = data.transpose()
coeffMatrix  = data_t.iloc[0:100,:]
```

```python
coeffMatrix = coeffMatrix[coeffMatrix.columns].astype(float)  # or int
col_names   = coeffMatrix.index
col_names   = list(map(lambda x: x.split('_'),col_names))
col_names   = list(map(lambda x : x[0],col_names))
coeffMatrix.index = col_names
```

```python
fig = sns.heatmap(coeffMatrix,vmin=-0.03, vmax=0.025,cmap="RdBu")
```

```python
fig.figure.savefig('coeff_matrix.png')
```

```python
pred = []
for i in range(data.shape[0]):
    pred.append(np.dot(data.iloc[i,101:201],data.iloc[i,0:100]) + data.iloc[i,101])
```

```python
plt.plot(data.iloc[:,201],pred,'ro')
plt.ylim(-0.2, 0.1)
plt.xlim(-0.2, 0.1)

z = np.polyfit(data.iloc[:,201], pred, 1)
p = np.poly1d(z)
plt.plot(data.iloc[:,201],p(data.iloc[:,201]),"r--",color="black")
plt.xlabel('Actual adjusted survival months')
plt.ylabel('Predicted adjusted survival months')
plt.savefig('predict.png')
```

**Correlation Coefficient**

```python
corr = np.corrcoef(data.iloc[:,201], pred)
corr = corr[0,0]
```
